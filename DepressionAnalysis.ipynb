### Data Cleaning and Preparation

```python
dataset['Gender'] = dataset['Gender'].map({'Male': 1, 'Female': 0})
dataset['Dietary Habits'] = dataset['Dietary Habits'].map({'Moderate' : 1, 'Unhealthy' : 0, 'Healthy' : 2})
dataset['Have you ever had suicidal thoughts ?'] = dataset['Have you ever had suicidal thoughts ?'].map({'Yes': 1, 'No': 0})
dataset['Family History of Mental Illness'] = dataset['Family History of Mental Illness'].map({'Yes': 1, 'No': 0})
dataset['Sleep Duration'] = dataset['Sleep Duration'].map({'7-8 hours': 2, 'Less than 5 hours': 0, '5-6 hours': 1, 'More than 8 hours' : 3})
dataset['Depression'] = dataset['Depression'].map({'Yes': 1, 'No': 0})
dataset.head()
```

```python
X = dataset.iloc[:,0:-1]
y = dataset.iloc[:,-1]
X.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,stratify = y, random_state = 1)
print(X.shape, X_train.shape, X_test.shape, X_train.ndim)
print(y.shape, y_train.shape, y_test.shape, y_train.ndim)
```

```python
depression_bot = LogisticRegression()
depression_bot.fit(X_train, y_train)
```

```python
y_pred = depression_bot.predict(X_test)
X_test.head()
```

```python
accuracy_score(y_test, y_pred)
```
